# A.Flow Control
:-Flow control describes the order in which statement will be executed at runtime.

Type of Flow cintrol
:-Python flow control statements are mainly divided into three types: 

           (1) Conditional Statements (Decision Making)             
           (2) Looping Statements (Iteration) 
           (3) Jump Statements (Control Transfer)


# (1)Conditional Statement

:- There are 3 main conditional statements (conditions):

       i)if
       ii)if – else
       iii) if – elif – else

    1) An if statement is used for a single condition.

         syntax
    if condition:
       statement 
       
     (2) If two conditions are to be executed within a single statement, the if-else structure is used.

        syntax 
        if condition :
           statement 
        else:
           statement
           
      (3) When there are multiple conditions, we use else if statement.

       syntax
       if condition:
           statement
        elif condition:
           statement
        else:
           statement

#Today, I built a simple “Choose Your Own Adventure” text-based game using Python. This project applies core programming concepts such as conditional statements, user input handling, and string manipulation that I learned during the lesson. The game allows users to make choices that lead to different outcomes.

I can use the lower() function to convert any string into lowercase.

Alternatively, you can use the logical operator or to check different spellings of the user’s choice, such as Right, right, or RIGHT

```python
print("Welcome to Treasure Island.")
print("Your mission is to find the treasure.")

choice1 = input(
    'You\'re at a crossroad. Where do you want to go? Type "left" or "right": '
).lower()

if choice1 == "left":
    choice2 = input(
        'You\'ve come to a lake. '
        'There is an island in the middle of the lake. '
        'Type "wait" to wait for a boat. '
        'Type "swim" to swim across: '
    ).lower()

    if choice2 == "wait":
        choice3 = input(
            "You arrive at the island unharmed. "
            "There is a house with 3 doors: one red, one yellow, and one blue. "
            "Which colour do you choose? "
        ).lower()

        if choice3 == "red":
            print("It's a room full of fire. Game Over.")
        elif choice3 == "yellow":
            print("You found the treasure. You win! ")
        elif choice3 == "blue":
            print("You enter a room of beasts. Game Over.")
        else:
            print("You chose a door that doesn't exist. Game Over.")

    else:
        print("You got attacked by an angry trout. Game Over.")

else:
    print("You fell into a hole. Game Over.")

    
```
