```python
#For example 
is_sunny = True   # Correct
is_raining = False  # Correct
```

# 7.List
Definition :- A list in Python is a collection data type used to store multiple values in a single variable.

Key Features of List:-

    (i)   Written using square brackets [ ].
    (ii)  Can store different data types (int, float, string, boolean, etc.
    (iii) Ordered → items have fixed positions (index).
    (iv)  Mutable → values can be changed.
    (v)   Allows duplicate values.

```python
my_list = [10, 20, 30, "apple", True, 2.5]
print(my_list)
print(type(my_list))
```

# 8.Dictionary
Definition :-
A dictionary in Python is a collection data type used to store data in key–value pairs.

Key Features of Dictionary:-

     (i)  Written using curly braces { }
    (ii)  Stores data as key : value
    (iii) Unordered (in older Python versions), insertion-ordered in modern Python
    (iv)  Mutable → values can be changed
    (v)   Keys must be unique
    (vi)  Values can be of any data type

```python
# For example :-
student = {
    "name": "Nisha",
    "age": 21,
    "course": "Python",
    "is_active": True
}

print(student)
print(type(student))
```

# 3.Type conversion / Type Casting

:- you can convert data into different data type using function 
e.g

     float()
     int()

:- Python supports a wide variety of built-in functions such as int(), float(), str(), ord(), hex(), tuple(), set(), list(), and dict() that are used for type casting.

:- Their are two type of type casting such as :- 

      (1)Explicit Type Casting:- The programmer manually converts one data type into another using functions like int(), float(), str().

      (2)Implicity type casting :- :-Implicit type casting is when Python   automatically converts one data type into another without any instruction from the programmer.



```python
# extplicit typecasting:
a= "23"
b= "32"
c= int(a) +int(b)
print(c)
```

```python
# Implicit type casting 
c=1.9                 #<class "int">
d=8                   #<class "float">
print(c+d)            #9.9
                      #<class "float">
```
