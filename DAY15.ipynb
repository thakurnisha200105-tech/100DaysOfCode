# OOPs (Object-Oriented Programming)

OOPs is a programming approach that is based on classes and objects.
In simple words, OOPs is a way to represent real-world entities in programming.

# Main Concepts of OOPs

There are two important terms in OOPs:

1. Object

An object is any real-world entity.
For example: mouse, keyboard, pen, car, student — all these are objects.
Each object has its own data and behavior.

2. Class

A class is a blueprint or template used to create objects.

Example:
Imagine a car factory.
The factory uses one blueprint to manufacture many cars like X, Y, and Z.
Even though all cars are different units, they have the same features and functions because they are created from the same blueprint.

# OOP Example (Chair & Table)

    A chair is a type of furniture that has four legs and a back support.

                 That means:

      Chair and Table are related
      Both have 4 legs
      Chair has an extra feature (back support)

                This is how OOP works:

      Chair = Class (blueprint)
      Real chair in your house = Object
      Legs, seat, back support = Attributes

# Exter parameter pass


```python
class Faculty:
    def __init__(self, a, b, c):  # constructor
        self.f_id = a
        self.name = b
        self.salary = c

    def display(self):
        print(f"Faculty ID: {self.f_id}")
        print(f"Faculty Name: {self.name}")
        print(f"Faculty Salary: {self.salary}")


a = Faculty(1, "Nisha Thakur", 2000)
a.display()
```

# oops (Class and instance attributes)


```python
class Student:
    university = "Amity university "      # class attribute

    def __init__(self, studentid, studentname):
        self.studentid = studentid      # instance attribute
        self.studentname = studentname  # instance attribute


# creating objects
c1 = Student(111, "Shivam")
c2 = Student(222, "Nisha")
c3 = Student(333, "Umesh")

print(c1.studentid)       # 111
print(c1.studentname)     # Ram

print(Student.university)     # amity university ----class attribute calls 
print(c1.university)          # amity university 

c1.studentid = 555            # “These two lines mean that we are changing the student’s ID for object c1,
                              #  and we are also updating the school (university) name for the entire class.”
Student.university = "AIIT"
```
